# 1- Problem Statement

#### - You own an ice cream business and you want to create a model to predict daily revenue in dollars based on temperature (degC).
#### - You decided to build a simple Artificial Neural Network to solve this Problem.
#### - Data set:
	- Input (X): Outside Air Temperature
	- Output (y): Overall daily revenue generated in Dollars
#### - This is A Regression Example

# 0- IMPORT LIBRARIES

# 2- VISUALIZE DATASET

```python
sns.scatterplot(sales_df['Temperature'], sales_df['Revenue'])
```

# 3- CREATE TESTING AND TRAINING DATASET

```python
X_train = sales_df['Temperature']
y_train = sales_df['Revenue']
```

# 4- BUILD AND TRAIN THE MODEL

```python
X_train.shape
```

```python
y_train.shape
```

```python
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
model = Sequential()
model.add(Dense(units = 1, input_shape = [1]))
```

```python
model.summary()
```

```python
model.compile(optimizer=Adam(0.1), loss = 'mean_squared_error' )
```

```python
epochs = model.fit(X_train, y_train, epochs = 1000)
```

# 5- EVALUATE THE MODEL

```python
epochs.history.keys()
```

```python
plt.plot(epochs.history['loss'])
plt.title('Model Loss Progress During Training')
plt.xlabel('Epoch')
plt.ylabel('Training Loss')
plt.legend(['Training Loss'])
```

```python
model.get_weights()
```

```python
# test the model
TEMP = 5
revenue = model.predict([TEMP])
print('Revenue Prediction Using Trained ANN =', revenue)
```

```python
plt.scatter(X_train, y_train, color = 'gray')
plt.plot(X_train, model.predict(X_train), color = 'red')
plt.ylabel('Revenue [dollars]')
plt.xlabel('Temperature [degC]')
plt.title('Revenue Generated vs. Temerature @Ice Cream Stand')
```

# 6- CONFIRM USING REGRESSION KIT IN SCIKIT LEARN

```python
X_train = X_train.values.reshape(-1, 1)
```

```python
X_train.shape
```

```python
y_train = y_train.values.reshape(-1, 1)
```

```python
y_train.shape
```

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression(fit_intercept = True)
regressor.fit(X_train, y_train)
```

```python
print('Linear Model Coeff(m)', regressor.coef_)
print('Linear Model Coeff(b)', regressor.intercept_)
```

```python
plt.scatter(X_train, y_train, color = 'gray')
plt.plot(X_train, regressor.predict(X_train), color = 'red')
plt.ylabel('Revenue [dollars]')
plt.xlabel('Temperature [degC]')
plt.title('Revenue Generated vs. Temerature @Ice Cream Stand')
```
